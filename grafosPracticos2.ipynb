<h3 style="color: #007bff; margin-bottom: 5px;"> <strong> ¿Qué son los Grafos? </strong> </h3>

Un grafo es una estructura matemática que se utiliza para modelar pares de objetos entre los cuales existen relaciones. Consiste en nodos, también llamados vértices, que representan los objetos, y aristas, que son las conexiones entre los vértices. Estos pueden representar redes de diversos tipos, como sociales, informáticas o de transporte, facilitando el estudio y la visualización de las relaciones complejas entre los elementos que componen estas redes.

<h3 style="color: #007bff; margin-bottom: 5px;"><strong>Aplicaciones Prácticas</strong></h3>

* Utilizados para encontrar la ruta más corta entre dos puntos, donde los nodos son ubicaciones y las aristas son caminos o carreteras.
  
* En bases de datos, especialmente en bases de datos de grafos, para modelar relaciones dinámicas entre datos.

<h3 style="color: #007bff; margin-bottom: 5px; text-align: center;"><strong>DBSCAN</strong></h3>
<p>"Density-Based Spatial Clustering of Applications with Noise", es un algoritmo de clustering que identifica grupos basados en la densidad de puntos en un espacio. Funciona con dos parámetros:</p>
<ul>
  <li><strong>eps</strong> (epsilon): Este parámetro define el radio de vecindad alrededor de cada punto. Puntos que están dentro de esta distancia son considerados vecinos.</li>
  <li><strong>minPts</strong> (minimum points): Es el número mínimo de puntos requeridos para formar un cluster densamente poblado. Un punto necesita tener al menos <em>minPts</em> vecinos dentro de su radio <em>eps</em> para ser considerado un punto central, o núcleo, del cluster.</li>
</ul>

<h3 style="color: #007bff; margin-bottom: 5px;"><strong>Carga de datos</strong></h3>

```python
from ClassGraph import *
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
grafoTabla = pd.read_csv("mapadistancias.csv").drop("Unnamed: 0", axis=1)
grafoTabla.head()
```

<h3 style="color: #007bff; margin-bottom: 5px;"><strong>Creación de Grafo</strong></h3>

```python
g = Graph()

vertex_map = {} 
for index, row in grafoTabla.iterrows():
    cluster_vertex = vertex_map.get(row['cluster'])
    if cluster_vertex is None:
        cluster_vertex = g.insert_vertex(row['cluster'])
        vertex_map[row['cluster']] = cluster_vertex

    cai_vertex = vertex_map.get(row['Cai'])
    if cai_vertex is None:
        cai_vertex = g.insert_vertex(row['Cai'])
        vertex_map[row['Cai']] = cai_vertex

    g.insert_edge(cluster_vertex, cai_vertex, row['distancia'])
```

```python
print(f" el número total de vértices es {g.vertex_count()}")
print(f" el número total de aristas es {g.edge_count()}")
```

<h3 style="color: #007bff; margin-bottom: 5px;"><strong>Grafo Networkx</strong></h3>

```python
import networkx as nx
G = nx.Graph()

def ordenar(diccionario): return dict(list(sorted(diccionario.items(), key=lambda item: item[1], reverse=True))[:5])
def invertido(diccionario): return dict(list(sorted(diccionario.items(), key=lambda item: item[1], reverse=False))[:5])

# Agregar nodos para cada cluster y cada Cai
clusters = grafoTabla['cluster'].unique()
for cluster in clusters:
    G.add_node(f'Cluster {cluster}', tipo='cluster')

for index, row in grafoTabla.iterrows():
    G.add_node(row['Cai'], tipo='cai', criminalidad=row['CriminalidadLocalidad'])
    # Conectar cada Cai a su cluster correspondiente
    G.add_edge(f'Cluster {row["cluster"]}', row['Cai'], distancia=row['distancia'], tiempo=row['tiempo'])
```

```python
print(G.number_of_nodes())
(G.number_of_edges())
#list(G.nodes(data=True))
```

```python
list(G.neighbors('Cai Dorado '))
```

<h3 style="color: #007bff; margin-bottom: 5px;"><strong>Centralidad del Grado</strong></h3>
Esta métrica mide la importancia o influencia de un nodo dentro de la red, basada en el número de conexiones (aristas) que tiene.

```python
degree_centrality = nx.degree_centrality(G)
ordenar(degree_centrality)
```

<h3 style="color: #007bff; margin-bottom: 5px;"><strong>Centralidad por el camino</strong></h3>
 La cercanía se mide en términos de la longitud del camino más corto promedio entre el nodo en cuestión y todos los otros nodos del grafo.

```python
closeness_centrality = nx.closeness_centrality(G)
invertido(closeness_centrality)
```

<h3 style="color: #007bff; margin-bottom: 5px;"><strong>Distribución del grado</strong></h3>
genera un histograma de la distribución de grados en un grafo G

```python
degree_distribution = nx.degree_histogram(G)
degree_distribution
```

```python
degrees = range(len(degree_distribution))

plt.figure(figsize=(10, 6))
plt.bar(degrees, degree_distribution, color='g')
plt.title('Distribución de Grados del Grafo')
plt.xlabel('Grado')
plt.ylabel('Número de Nodos')

# Establecer los ticks del eje x manualmente para mostrar enteros
plt.xticks(range(len(degrees)), [str(int(d)) for d in degrees])

plt.show()
```

<h3 style="color: #007bff; margin-bottom: 5px;"><strong>Emparejamientos óptimos</strong></h3>
En el contexto de grafos ponderados, un emparejamiento de peso máximo es aquel donde la suma de los pesos de las aristas incluidas en el emparejamiento es máxima.

```python
emparejamiento = nx.algorithms.matching.max_weight_matching(G, maxcardinality=False)
list(emparejamiento)[:5]
```

<h3 style="color: #007bff; margin-bottom: 5px;"><strong>Centralidad de intermediación</strong></h3>
Evalúa cuántas veces un nodo actúa como puente a lo largo del camino más corto entre otros pares de nodos.

```python
import matplotlib.pyplot as plt

betweenness = nx.betweenness_centrality(G)

print("Centralidad de Intermediación de cada nodo:")
for node, centrality in betweenness.items():
    print(f"Nodo {node}: {centrality:.2f}")
```

```python
# Visualizar el grafo
pos = nx.spring_layout(G)  # Para la visualización
nx.draw(G, pos, with_labels=False, node_color='lightblue', edge_color='#AAAAAA', node_size=[v * 100 for v in betweenness.values()])
plt.show()
```
